#From the parameter column we will keep the EV sales.

```python
#drop the historical column 
def clean_columns(df):
    df.columns = [column.replace(" ", "_").lower() for column in df.columns]
```

```python
clean_columns(ev)
```

```python
ev
```

```python
def drop_category (df):
    del df['category']
```

```python
drop_category(ev)
```

```python
ev
```

```python
#drop the world row
def drop_world(df):
    df.drop(df[df['region'] == 'World'].index, inplace=True)
```

```python
drop_world (ev)
```

```python
ev
```

```python
def keep_ev_sales(df):
    return df[df['parameter'] == 'EV sales']
```

```python
keep_ev_sales(ev)
```
